```python
def process_string(text):
    reversed_str = text[::-1]
    second_chars = text[1::2]
    return reversed_str, second_chars

s = input("Enter a string: ")

rev, sec = process_string(s)

print("Reversed string:", rev)
print("Every 2nd character:", sec)
print("just for commmiting update")
```

```python
# Heading: Squares of even numbers 2
def square_even(nums):
    return [x**2 for x in nums if x % 2 == 0]

print(square_even([1,2,3,4,5,6]))
print("just for commmiting update")
```

```python
#Heading: Character frequency 3
def char_frequency(text):
    freq = {}
    for ch in text:
        freq[ch] = freq.get(ch, 0) + 1
    return freq

print(char_frequency("hussain"))
```

```python
#Heading: Indices of odd numbers 4
def odd_indices(nums):
    return [i for i, x in enumerate(nums) if x % 2 != 0]

print(odd_indices([10, 15, 20, 25, 30]))
```

```python
#Heading: Rotate list right by k 5
def rotate_right(lst, k):
    k = k % len(lst)
    return lst[-k:] + lst[:-k]

print(rotate_right([1,2,3,4,5], 2))
```

```python
#Heading: Two lists 6
def make_dict(keys, values):
    return dict(zip(keys, values))

print(make_dict(['a','b','c'], [1,2,3]))
```

```python
#Heading: iter() and next() with exception 7
def manual_iter(lst):
    it = iter(lst)
    while True:
        try:
            print(next(it))
        except StopIteration:
            print("End of list")
            break

manual_iter([1,2,3]) 
```

```python
#heading: Flatten nested list 8
def flatten(nested):
    return [item for sublist in nested for item in sublist]

print(flatten([[1,2],[3,4],[5]]))
```

```python
#Heading: Palindrome check (slicing) 9 
def is_palindrome(s):
    return s == s[::-1]

print(is_palindrome("madam"))
```

```python
#Heading: Second largest (without sorting) 10
def second_largest(nums):
    first = second = float('-inf')
    
    for n in nums:
        if n > first:
            second = first
            first = n
        elif n > second and n != first:
            second = n
    
    return second

print(second_largest([10, 20, 4, 45, 99]))
```
